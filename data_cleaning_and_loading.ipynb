# MLB

```python
#import relevant libraries
import pandas as pd
```

```python
csv = pd.read_csv("2022 MLB Salaries.csv")
```

```python
csv
```

```python
mlbDF = csv[["Player","Team", "Salary"]]
mlbDF
```

```python
mlbDF.info()
```

```python
#get rid of '$' and ',' within Salaries, and then convert type to integer
mlbDF["Salary"]= mlbDF["Salary"].str.replace('$','',regex=True).str.replace(',','',regex=True).astype(int)
```

# NFL

```python
#Sum up player salaries for each team
mlbTeamDF = mlbDF.groupby("Team").sum()
mlbTeamDF = mlbTeamDF.reset_index()
mlbTeamDF
```

```python
mlbTeamDF["Sport"] = ["MLB"]*len(mlbTeamDF)
mlbTeamDF
```

```python
csv_nfl = pd.read_csv("TeamSal22.csv")
```

```python
csv_nfl
```

```python
csv_nfl["sport"] = ["NFL"]*len(csv_nfl)
csv_nfl
```

```python
csv_nfl_upper = csv_nfl.rename(columns={"team":"Team", "salary":"Salary", "sport":"Sport"})
csv_nfl_upper
```

# MLB and NFL

```python
cap_df = pd.concat([mlbTeamDF, csv_nfl_upper])
cap_df
```
