```python
#import relevant libraries
import pandas as pd
```

```python
csv = pd.read_csv("2022 MLB Salaries.csv")
```

```python
csv
```

```python
mlbDF = csv[["Player","Team", "Salary"]]
mlbDF
```

```python
mlbDF.info()
```

```python
#get rid of '$' and ',' within Salaries, and then convert type to integer
mlbDF["Salary"]= mlbDF["Salary"].str.replace('$','',regex=True).str.replace(',','',regex=True).astype(int)
```

```python
#Sum up player salaries for each team
mlbTeamDF = mlbDF.groupby("Team").sum()
mlbTeamDF = mlbTeamDF.reset_index()
mlbTeamDF
```
